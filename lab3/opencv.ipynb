```python
import cv2
from matplotlib import pyplot as plt
image = cv2.imread("img.jpg", 1)
plt.axis("off")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
width = int(image.shape[1] * 50 / 100)
height = int(image.shape[0] * 50 / 100)
dim = (width, height)
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
plt.axis("off")
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.show()

print("raw image size:")
print(image.shape[1], image.shape[0])

print("resized image:")
print(resized.shape[1], resized.shape[0])
```

```python
(h, w, d) = image.shape
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 270, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))

plt.axis("off")
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.show()

print("rotated to 270 degree")
```

```python
print(center)
output = image.copy()
cv2.rectangle(output, (400, 400), (850, 700), (0, 255, 255), 15)

plt.axis("off")
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
output2 = output.copy()
cv2.putText(output2, "just text under rectangle", (100, 800),cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 3, (0, 255, 255), 5) 

plt.axis("off")
plt.imshow(cv2.cvtColor(output2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imsave("img2.png", cv2.cvtColor(output2, cv2.COLOR_BGR2RGB))
```
