```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
iris_dataset = datasets.load_iris()

X = iris_dataset.data
Y = iris_dataset.target

x_index = 0
y_index = 1


plt.figure(figsize=(5, 4))
plt.scatter(iris_dataset.data[:,x_index], iris_dataset.data[:, y_index], c=iris_dataset.target)

plt.xlabel(iris_dataset.feature_names[x_index])
plt.ylabel(iris_dataset.feature_names[y_index])

plt.tight_layout()
plt.show()
```

```python
trainX = X[:int(len(X)*0.7)]
trainY = Y[:int(len(Y)*0.7)]

testX = X[int(len(X) * 0.7):]
testY = Y[int(len(Y) * 0.7):]
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
dddd = knn.fit(trainX, trainY)
# print(dddd)
# knn.score()
predicted = dddd.predict(testX)
```

```python
accuracy = knn.score(testX, testY)
print(accuracy)
```
